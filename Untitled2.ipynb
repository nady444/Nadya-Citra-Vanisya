```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
file_name = list(uploaded.keys())[0]  # Ambil nama file yang diunggah
df = pd.read_csv(file_name)  # Baca file CSV

# Set style
sns.set(style="whitegrid")

# 1. Distribusi IMDb Score
plt.figure(figsize=(8, 5))
sns.histplot(df['imdb_score'], bins=20, kde=True, color="blue")
plt.title("Distribusi IMDb Score")
plt.xlabel("IMDb Score")
plt.ylabel("Jumlah Film")
plt.show()

"""
Penjelasan:
Grafik ini menunjukkan distribusi skor IMDb film dalam dataset. Histogram ini menggunakan 20 bin,
dan garis KDE (Kernel Density Estimation) membantu melihat pola distribusi skor film secara halus.
"""

# 2. Scatter plot Budget vs Gross
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df['budget'], y=df['gross'], alpha=0.6, color="red")
plt.title("Budget vs Gross (Pendapatan)")
plt.xlabel("Budget (USD)")
plt.ylabel("Gross (Pendapatan USD)")
plt.show()

"""
Penjelasan:
Grafik ini menunjukkan hubungan antara budget film dengan pendapatan kotor yang dihasilkan.
Scatter plot ini dapat membantu melihat apakah ada hubungan positif antara budget dan pendapatan.
"""

# 3. Scatter plot Jumlah Vote vs IMDb Score
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df['num_voted_users'], y=df['imdb_score'], alpha=0.6, color="green")
plt.title("Jumlah Vote vs IMDb Score")
plt.xlabel("Jumlah Vote")
plt.ylabel("IMDb Score")
plt.xscale("log")
plt.show()

"""
Penjelasan:
Scatter plot ini menunjukkan korelasi antara jumlah vote pengguna dan skor IMDb.
Sumbu X menggunakan skala logaritmik untuk menyesuaikan distribusi data yang cenderung memiliki nilai ekstrem.
"""

# 4. Facebook Likes vs IMDb Score
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df['movie_facebook_likes'], y=df['imdb_score'], alpha=0.6, color="purple")
plt.title("Facebook Likes vs IMDb Score")
plt.xlabel("Movie Facebook Likes")
plt.ylabel("IMDb Score")
plt.xscale("log")
plt.show()

"""
Penjelasan:
Scatter plot ini menunjukkan hubungan antara jumlah likes di Facebook dan skor IMDb.
Sumbu X juga menggunakan skala log untuk menangani data dengan rentang yang luas.
"""

# 5. Jumlah Review Pengguna vs Kritikus
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df['num_user_for_reviews'], y=df['num_critic_for_reviews'], alpha=0.6, color="orange")
plt.title("Jumlah Review Pengguna vs Kritikus")
plt.xlabel("Jumlah Review Pengguna")
plt.ylabel("Jumlah Review Kritikus")
plt.show()

"""
Penjelasan:
Scatter plot ini membandingkan jumlah review dari pengguna dan kritikus.
Jika terdapat korelasi positif, maka jumlah review pengguna yang tinggi cenderung sejalan dengan review kritikus yang lebih banyak.
"""

# 6. Distribusi Film berdasarkan Negara
plt.figure(figsize=(8, 5))
sns.barplot(x=df['country'].value_counts().index, y=df['country'].value_counts().values, palette="Purples_r", hue=df['country'].value_counts().index, legend=False)
plt.xticks(rotation=45)
plt.title("Distribusi Film berdasarkan Negara")
plt.xlabel("Negara")
plt.ylabel("Jumlah Film")
plt.show()

"""
Penjelasan:
Grafik ini menunjukkan jumlah film berdasarkan negara asal dalam dataset.
Negara dengan jumlah film terbanyak akan terlihat lebih tinggi dalam bar chart ini.
Rotasi label sumbu X digunakan untuk meningkatkan keterbacaan.
"""
```
